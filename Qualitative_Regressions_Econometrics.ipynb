```python
df = pd.read_csv('casa.csv', sep=';')
```

```python
any_nulls = df.isnull().any().any()
print(f"valores nulos? {any_nulls}")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import logit, probit, ols



# Regress√£o Linear (MPL)
X = sm.add_constant(df['renda'])
y = df['casa']
model_mpl = sm.OLS(y, X).fit()
print(model_mpl.summary())

# Regress√£o Logit
model_logit = logit('casa ~ renda', data=df).fit()
print(model_logit.summary())

# Regress√£o Probit
model_probit = probit('casa ~ renda', data=df).fit()
print(model_probit.summary())
```
