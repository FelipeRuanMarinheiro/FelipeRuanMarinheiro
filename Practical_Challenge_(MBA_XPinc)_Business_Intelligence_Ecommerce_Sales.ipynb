```python
#Preenchendo os valores de produto, pela lógica de negócios.

# Função para calcular o valor do produto
def calcular_valor_produto(row):
    if pd.isna(row['Valor do produto']):
        return row['Valor da venda'] / row['Quantidade unitária']
    else:
        return row['Valor do produto']

# Preencher os valores faltantes na coluna "Valor do produto"
df['Valor do produto'] = df.apply(calcular_valor_produto, axis=1)
```

```python
#Preenchendo os valores da venda, pela lógica de negócios.

# Função para calcular o valor da venda
def calcular_valor_venda(row):
    if pd.isna(row['Valor da venda']):
        return row['Valor do produto'] * row['Quantidade unitária']
    else:
        return row['Valor da venda']

# Preencher os valores faltantes na coluna "Valor do produto"
df['Valor da venda'] = df.apply(calcular_valor_venda, axis=1)
```

```python
#identificando dados faltantes
df.isnull().sum()
```

```python
#Qual é a média do valor das vendas de todos os produtos destinados ao Estado de Minas Gerais (MG)?
df.groupby('Localização')['Valor da venda'].mean()
```

```python
#Qual produto possui o maior número de vendas?  Avalie o produto que mais teve vendas.
df.groupby('Produto vendido')['Quantidade unitária'].sum()
```

```python
#Qual é o desvio padrão da quantidade unitária para todos os produtos vendidos no Estado de Pernambuco (PE)?
df_pe = df[df['Localização'] == 'PE']
desvio_padrao = df_pe['Quantidade unitária'].std()
print(desvio_padrao)
```

```python
#Qual é o valor máximo de venda registrado para os produtos comercializados no Estado de Santa Catarina (SC)?
df_sc = df[df['Localização'] == 'SC']
max = df_sc['Valor da venda'].max()
print(max)
```

```python
#Faça o somatório do valor total das vendas de todos os produtos identificados como "Produto A". Em seguida, forneça o resultado do somatório obtido.
df_prod_a = df[df['Produto vendido'] == 'Produto A']
sum = df_prod_a['Valor da venda'].sum()
print(sum)
```

```python
#Qual é a média da quantidade unitária vendida para o Produto B no Estado de Pernambuco (PE)?
# Filtrar o DataFrame para incluir apenas as linhas onde o produto vendido é "Produto B" e a localização é "PE"
df_pe_produto_b = df[(df['Produto vendido'] == 'Produto B') & (df['Localização'] == 'PE')]

# Calcular a média da quantidade unitária
media_quantidade_unitaria = df_pe_produto_b['Quantidade unitária'].mean()

print(media_quantidade_unitaria)
```

```python
#Qual é a soma do valor da venda para todos os produtos vendidos para os estados de MG e SP?
# Filtrar o DataFrame para incluir apenas as linhas onde a localização é "MG" ou "SP"
df_MG_sp = df[(df['Localização'] == 'MG') | (df['Localização'] == 'SP')]



# Calcular a soma do valor da venda
soma_valor_venda_MG_SP = df_MG_sp['Valor da venda'].sum()

print(soma_valor_venda_MG_SP)
```

```python
#Calcule a média da coluna "Valor do produto" para todos os produtos que estão disponíveis nos estados de AL, BA, CE, MA, PB, PE, PI, RN, SE. Qual é o valor resultante?
# Lista dos estados
estados = ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']

# Filtrar o DataFrame para incluir apenas as linhas onde a localização está em estados
df_estados = df[df['Localização'].isin(estados)]

# Calcular a média da coluna "Valor do produto"
media_valor_produto = df_estados['Valor do produto'].mean()

print(media_valor_produto)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtendo os valores únicos da coluna 'Localização'
localizacoes = sorted(df['Localização'].unique())


# Definindo o tamanho dos gráficos
plt.figure(figsize=(10, 7 * len(localizacoes)))

# Iterando sobre cada valor único da coluna 'Localização' e criando um gráfico para cada um
for i, localizacao in enumerate(localizacoes, 1):
    plt.subplot(len(localizacoes), 1, i)
    plt.title(f'Dispersão - {localizacao}')
    sns.scatterplot(data=df[df['Localização'] == localizacao],
                    x='Data da compra',
                    y='Valor da venda'
                    )

    # Adicionando uma linha horizontal na média dos valores de 'Valor da venda'
    media_valor_venda = df[df['Localização'] == localizacao]['Valor da venda'].mean()
    plt.axhline(y=media_valor_venda, color='r', linestyle='--', label=f'Média: {media_valor_venda:.2f}')
    plt.xlabel(' ')
    plt.ylabel('R$')
    plt.legend()

plt.tight_layout()
plt.show()


```
