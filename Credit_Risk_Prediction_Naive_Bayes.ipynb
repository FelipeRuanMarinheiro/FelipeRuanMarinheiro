##Importanto libs

```python
!pip install scikit-learn
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_excel('riscos.xlsx')
df
```

```python
#Selecionando colunas para criar indicador
col_sel=['renda', 'divida', 'fiador', 'historico']
```

```python
#Criando identificador para cada uma das variaveis categoricas
LE = LabelEncoder()

#Criando coluna baseada no encoder
for coluna in col_sel:
  df[f'id_{coluna}'] = LE.fit_transform(df[coluna])

df
```

```python
#organizando colunas:
df = df[['id_renda','renda', 'id_divida', 'divida', 'id_fiador', 'fiador', 'id_historico', 'historico', 'risco']]
df
```

```python
#Selecionando apenas colunas transformadas para usar no modelo:
col_mod = ['id_renda', 'id_divida', 'id_fiador', 'id_historico']
df[col_mod]
```

```python
#Separando features do target
X_dados = df[col_mod]
y_dados = df['risco']
```

```python
#Criação do modelo
model = GaussianNB()
model.fit(X_dados.values, y_dados.values)
```

```python
model.classes_ #visualização das classes
```

```python
model.class_count_ #contagem das classes, 2 de alto, 4 de baixo, 4 de moderado...
```

```python
model.class_prior_ #frequencias das classes
```

```python
model.score(X_dados.values, y_dados.values)
```

```python
model.predict(X=X_dados.values)
```

```python
df['predict'] =  model.predict(X_dados.values)
df
```

```python
dados_cliente = [[0, 1, 1, 2]]
model.predict(dados_cliente)
```
